```python
def cal_fit_func(x,a,b):
    return a*x+b
```

```python
eV_points = [22.59,363.758,387.461]
```

```python
fit_pars = []
x = np.arange(0,600)
for i in range(len(data)):
    adc = [xray_data['p4'][i],ecap_data['p1'][i],ecap_data['p4'][i]]
    a, b = np.polyfit(eV_points, adc, deg=1)
    fit_pars.append([a,b])
    
    plt.scatter(eV_points,adc,marker = '.',label = '%d'%i)
    plt.plot(x,cal_fit_func(x,fit_pars[i][0],fit_pars[i][1]),linestyle = ':')
plt.legend()
plt.show()
```
